```python
# data science
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# regression models
from sklearn.linear_model import LinearRegression # for linear regression
```

```python
dataset = pd.read_csv('vivienda.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
# Rellenamos con la Mediana, con una funcion anidada, le dice hace el calculo de la mediana y rellena
df = pd.get_dummies(dataset)
df=df.fillna(df.median())
df
```

```python
df.isnull().sum()
```

```python
X = df.drop('TotalValue', axis = 1)
y = df['TotalValue']
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
print("Shape X_train: ",X_train.shape)
print("Shape X_test: ", X_test.shape)
print("Shape y_train: ",y_train.shape)
print("Shape y_test: ",y_test.shape)
```

```python
# Creating the linear regressor
regressor_linear = LinearRegression()
regressor_linear.fit(X_train, y_train)
```

```python
# Predict:
y_pred = regressor_linear.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Calculamos el error cuadrÃ¡tico medio y el estadÃ­stico R^2
print('Error cuadrÃ¡tico medio: %.2f' % mean_squared_error(y_test, y_pred))
#print('RMSE: %.2f' % (mean_squared_error(y_test, y_pred, squared=False)))
print('MAE: %.2f' % mean_absolute_error(y_test, y_pred))
print('R_2: %.2f' % r2_score(y_test, y_pred))
```

Con regresiÃ³n lineal nos da R_2=1, esto implica eficiencia del 100% del modelo.

IMPLEMENTAREMOS ALGORITMOS DE ENSAMBLE BAGGING Y BOOSTING

Random Forest Regressor -- Bagging

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Instanciamos el modelo
rfr = RandomForestRegressor(n_estimators=25)
```

```python
# Entrenamos:
rfr.fit(X_train,y_train)
```

```python
# PredicciÃ³n
y_pred = rfr.predict(X_test)
```

```python
# Calculamos el error cuadrÃ¡tico medio y el estadÃ­stico R^2
print('Error cuadrÃ¡tico medio: %.2f' % mean_squared_error(y_test, y_pred))
#print('RMSE: %.2f' % (mean_squared_error(y_test, y_pred, squared=False)))
print('MAE: %.2f' % mean_absolute_error(y_test, y_pred))
print('R_2: %.2f' % r2_score(y_test, y_pred))
```

Ada Boost Regressor -- Boosting

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
abr = AdaBoostRegressor(n_estimators=25)
```

```python
abr.fit(X_train,y_train)
```

```python
y_pred = abr.predict(X_test)
```

```python
# Calculamos el error cuadrÃ¡tico medio y el estadÃ­stico R^2
print('Error cuadrÃ¡tico medio: %.2f' % mean_squared_error(y_test, y_pred))
#print('RMSE: %.2f' % (mean_squared_error(y_test, y_pred, squared=False)))
print('MAE: %.2f' % mean_absolute_error(y_test, y_pred))
print('R_2: %.2f' % r2_score(y_test, y_pred))
```

ConclusiÃ³n: En este caso el boosting es menos eficiente que el bagging.
