```python
# data science
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# regression models
from sklearn.linear_model import LinearRegression # for linear regression
```

```python
dataset = pd.read_csv('vivienda.csv')
dataset.head()
```

```python
#Consultamos el tipo de dato de cada variable
dataset.info()
```

```python
dataset = pd.get_dummies(dataset)
dataset=dataset.fillna(dataset.median())
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.info()
```

```python
dataset['TaxDistrict_CITY OF GOODLETTSVILLE'] = dataset['TaxDistrict_CITY OF GOODLETTSVILLE'].astype(float)
dataset['TaxDistrict_GENERAL SERVICES DISTRICT'] = dataset['TaxDistrict_GENERAL SERVICES DISTRICT'].astype(float)
dataset['TaxDistrict_URBAN SERVICES DISTRICT'] = dataset['TaxDistrict_URBAN SERVICES DISTRICT'].astype(float)
dataset
```

```python
dataset['TaxDistrict_CITY OF GOODLETTSVILLE'].value_counts()
```

```python
dataset['TaxDistrict_GENERAL SERVICES DISTRICT'].value_counts()
```

```python
dataset['TaxDistrict_URBAN SERVICES DISTRICT'].value_counts()
```

```python
dataset.info()
```

```python
df=dataset
```

```python
X = df.drop('TotalValue',axis=1)
y = df['TotalValue']
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)
```

```python
# Creating the linear regressor
regressor_linear = LinearRegression()
regressor_linear.fit(X_train, y_train)
```

```python
# Predict:
y_pred = regressor_linear.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Calculamos el error cuadrÃ¡tico medio y el estadÃ­stico R^2
print('Error cuadrÃ¡tico medio: %.2f' % mean_squared_error(y_test, y_pred))
#print('RMSE: %.2f' % (mean_squared_error(y_test, y_pred,squared=False)))
print('MAE: %.2f' % mean_absolute_error(y_test, y_pred))
print('R_2: %.2f' % r2_score(y_test, y_pred))
```

CROSS VALIDATION SCORE

```python
from sklearn.model_selection import cross_val_score
# crear el modelo
model = LinearRegression()
scores = cross_val_score(model, X, y, cv=5, verbose=1,scoring='r2') # cv= 5
```

```python
print("%0.2f de r2 promedio con una desviaciÃ³n estandar de %0.2f" % (scores.mean(), scores.std()))
```
