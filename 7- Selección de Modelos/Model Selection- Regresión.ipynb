SELECCIÓN DE MODELOS - REGRESIÓN

```python
# data science
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('vivienda.csv')
df
```

```python
#Consultamos el tipo de dato de cada variable
df.info()
```

```python
df = pd.get_dummies(df)
df=df.fillna(df.median())
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['TaxDistrict_CITY OF GOODLETTSVILLE'] = df['TaxDistrict_CITY OF GOODLETTSVILLE'].astype(float)
df['TaxDistrict_GENERAL SERVICES DISTRICT'] = df['TaxDistrict_GENERAL SERVICES DISTRICT'].astype(float)
df['TaxDistrict_URBAN SERVICES DISTRICT'] = df['TaxDistrict_URBAN SERVICES DISTRICT'].astype(float)
df
```

```python
df.info()
```

```python
#df1=df['Acreage','LandValue', 'BuildingValue','TotalValue','Bedrooms','FullBath','HalfBath','TaxDistrict_CITY OF GOODLETTSVILLE','TaxDistrict_GENERAL SERVICES DISTRICT','TaxDistrict_URBAN SERVICES DISTRICT']
df1=df["Acreage"]
df2=df["LandValue"]
df3=df["BuildingValue"]
df4=df["TotalValue"]
df5=df["Bedrooms"]
df6=df["FullBath"]
df7=df["HalfBath"]
df8=df["TaxDistrict_CITY OF GOODLETTSVILLE"]
df9=df["TaxDistrict_GENERAL SERVICES DISTRICT"]
df10=df["TaxDistrict_URBAN SERVICES DISTRICT"]
df = pd.concat([df1, df2, df3, df5, df6, df7, df8, df9, df10, df4], axis=1)
df
```

```python
# Matriz Scatter
scatter_matrix(df, figsize = (8, 8), diagonal = 'kde')
plt.show()
```

```python
data = df.drop(['TotalValue'],axis=1)
data
```

```python
# Target
y1 = df['TotalValue']
y1
```

PYCARET

```python
from pycaret.utils import version
version()
```

```python
from sklearn.model_selection import train_test_split

seed = 42
# usamos la rutina train_test_split de Sklearn para dividir los datos en
# training y test (la variable random_state se usa para que siempre obtengamos la
# misma particion aleatoria)
X_train_val, X_test, y_train_val, y_test = train_test_split(data, y1, test_size=0.25, random_state=seed)

# ahora necesitamos partir X_train_val y y_train_val para sacarle el 10% que
# usaremos como validation
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.1, random_state=seed)

# imprimimos por pantalla los tamaños finales
print('=======================================================================================')
print('PARTICIÓN DATASET:')
print('')
print('Instancias de Entrenamiento:')
print('X_train: {}'.format(X_train.shape))
print('y_train: {}'.format(y_train.shape))
print(' ')
print('Instancias de Prueba:')
print('X_test: {}'.format(X_test.shape))
print('y_test: {}'.format(y_test.shape))
print(' ')
print('Instancias de Validación:')
print('X_val: {}'.format(X_val.shape))
print('y_val: {}'.format(y_val.shape))
print('=======================================================================================')
```

```python
#############################################################################
###  Creamos un Data Frame con los datos de Validacion para validar
#############################################################################

data_xval = pd.DataFrame(X_val, columns=data.columns)
data_xval.head()
```

```python
data_yval = pd.DataFrame(y_val)
data_yval.head()
```

```python
# Data Set para validar
data_unseen = pd.concat([data_xval,data_yval],axis=1)
data_unseen.head()
```

```python
# Data Set para entrenar en Pycaret:
data_py = pd.concat([X_train,y_train], axis=1)
data_py.head()
```
